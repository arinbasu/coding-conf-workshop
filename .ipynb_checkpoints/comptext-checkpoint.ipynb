## Features of tidy data and R
- tidy data has the following three properties
- Every observation or unit of observation has its own row
- Every feature of observation has its own column
- Every cell has only one information
- You can use tidy principles in conducting text data analysis
---
## Operations in tidyverse
- `%>%` stands for pipe
- This symbol is literally interpreted as 'then'
- `x %>% y` means x THEN y
- The above code says 
> load `mtcars` THEN show header data
---
## In text data analysis
- tidytext, each word or sentence will have its own row
- Each word in a corpus will have its features listed in columns
- We will summarise and operate on the word level data
- Alternatively, you can have sentence level data or 
- You can have character level data to work with
---
## Grammar of tidy data, part I
- `select()` selects columns
- `filter()` filters the rows based on conditions on columns
- `arrange()` rank orders variables in specified order
- `mutate()` create new variables from existing columns
- `group by() %>% summarise()` can summarise groups of data
---
## More with tidy data

```r
# grammar of tidy data
# select mpg: hp
mtcars1 <- mtcars %>%
  select(mpg:hp)

mtcars1 %>%
  head()
```

```r
# filter: select cars that have 6+ cylinders
six_plus <- mtcars1 %>%
    filter(cyl >= 4)
six_plus %>% head()
```

```r
# arrange() arrange the data set according to hp
six_plus %>% 
  arrange(hp) # arranges the data set in ascending order
```

```r
## arrange in descending order
six_plus %>%
  arrange(desc(hp))
```

```r
## group by cylinder size and give mean values of mpg
av_milage <- mtcars1 %>%
   group_by(cyl) %>%
   summarise(mean_mpg = mean(mpg, na.rm = T))

av_milage
 
```

```r
# create a new variable kilometers per liter based on miles per gallon
# based on 1 mile = 1.6 km, and 1 gallon = 3 liter
# therefore 1 mpg = (1.6 / 3 ) * kmpl
# hence, kmpl = mpg * (3 / 1.6)
mtcars2 <- mtcars1 %>%
    mutate(kmpl = mpg * (3 / 1.6))
mtcars2 %>%
    head()
```

```r
## tidy data part 2
# create a new variable economical based on median kmpl
mtcars3 <- mtcars2 %>%
  mutate(economical = (kmpl >= median(kmpl, na.rm = T)))
```

```r
# then tally the numbers
mtcars3 %>%
   count(economical)
```

```r
# Now that we know the distribution of economical, 
# are cars with more cylinders more economical?
cyl_econ <- mtcars3 %>%
    count(cyl, economical)
cyl_econ
```

```r
# the above table is too complex, how about spreading the data?
# We want to see True and false on columns next to cylinders

tbl_cyl_econ <- cyl_econ %>%
  spread(key = economical, value = n)
tbl_cyl_econ

## you can now take the tbl_cyl_econ and gather the FALSE and TRUE columns into a variable (say, 'econ')

tbl_cyl_gather <- tbl_cyl_econ %>%
   gather(`FALSE`, `TRUE`, key = econ, value = value) # note FALSE and TRUE are within backticks, 
                                                      # not boolean
tbl_cyl_gather
```

```r
## ggplot() is used for plotting
# let's plot engine horsepower (hp) with kmpl

mtcars3 %>%
  ggplot(aes(x = hp, y = kmpl)) +
  geom_point() +
  geom_smooth(method = "lm", 
              se = FALSE) +
  facet_wrap(~cyl) +
  theme_bw() +
  labs(title = "Horsepower vs milage by cylinder size",
      x = "Horspower",
      y = "Kilometers per litre")

```

```r
# Use these techniques for analysis of text data
# let's read some data 
# let's download text data from the complete works of Sherlock 
# Holmes (project gutenberg id: 1661)

sherlock <- gutenberg_download("1661")
```

```r
sherlock %>%
   head()
```

```r
# tidy sherlock for now
tidy_sherlock <- sherlock %>%
   unnest_tokens(word, text)
tidy_sherlock %>%
  head()
tidy_sherlock %>%
  dim()
```

# What do we see?
- We have exploded the entire corpus of text
- it has now 100,000+ words in it
- We could explode by sentences
- many of the words are commonplace words
- Words like 'the', 'of' are not useful for analyses
- these are called stopwords

```r
# what do stopwords look like
stopWords <- stopwords('english')
stopWords %>%
  head(10)
```

```r
# remove stopwords from the tidy_sherlock
data(stop_words) # we load a data base of stop words
tidy_sherlock <- tidy_sherlock %>%
   anti_join(stop_words)
tidy_sherlock %>%
  head()
```

```r
# now got tidied a little more
# we do not want words such as conan doyle sherlock holmes etc
unwanted_words <- c("sherlock", "holmes", "adventure", "adventures",
                   "conan", "doyle", "i", "ii", "iii", "iv",
                   "v", "vi", "vii", "viii", "ix", "x", "sir")
tidy_sherlock1 <- tidy_sherlock %>%
   filter(!word  %in% unwanted_words)

tidy_sherlock1 %>%
   head()
```

```r
# Let's do a word count and see how many words are there in 
# the entire corpus

sherlock_wc <- tidy_sherlock1 %>%
  count(word, sort = TRUE)

sherlock_wc %>%
  head()
```

```r
# Let's plot these
sherlock_wc %>%
  filter(n > 50) %>%
  mutate(word_reordered = reorder(word, n)) %>%
  ggplot(aes(x = word_reordered, y = n)) +
  geom_col() +
  xlab(NULL) +
  coord_flip()
```

```r
# let's see another way to read texts and analyse words
# in quanteda package, you will use corpus
# some corpus are already made available
# others you can read texts such as pdf documents using readtext
# here we read immigration data in The UK in 2010

uk_immig <- corpus(data_char_ukimmig2010)
# you can tokenise this corpus in quanteda in one step
# using document feature matrix

uk_immig1 <- uk_immig %>%
   dfm(remove = stopwords('english'),
      remove_punct = T,
      remove_numbers = T)
# this code has created a sparse matrix that we can now analyse

uk_immig1[, 1:5] # we will see 5 features
# total number of features
nfeat(uk_immig1) # total of 1489 features
# docfreq(uk_immig1)
#topfeatures(uk_immig1, groups = docnames(uk_immig1))
textplot_wordcloud(uk_immig1) # most common words in word cloud pattern
```

```r
## Sentiment analyses
# sentiment lexicons: afinn, bing, wordstat

afinn_list <- get_sentiments("afinn")
bing_list <- get_sentiments('bing')
bing_list %>%
  head(10)
```

```r
# let's do sentiment analysis of Sherlock Holmes stories
# and do sentiment analysis of British Election manifesto
# on immigration in 2010
# let's find all positive words in the Bing lexicon
bing_positive <- bing_list %>%
   filter(sentiment == "positive") # note the `==` sign

sherlock_positive <- tidy_sherlock1 %>%
   inner_join(bing_positive) %>%
   count(word, sort = TRUE)

sherlock_positive
```

```r
# create a line number variable for tidy_sherlock
tidy_sherlock1 %>%
 mutate(line_number = row_number())

tidy_sherlock1 %>%
  head()
```

```r
# let's see how per 50 lines the sentiments vary
# first create an index of every say 50 lines of text 
# in the entire corpus of text

sherlock_sentiment <- tidy_sherlock1 %>%
  inner_join(bing_list) %>%
  mutate(index = row_number() %/% 50)

# count the positive and negative sentiments

s_s <- sherlock_sentiment %>%
   count(index, sentiment)
# now spread the sentiments and create a sentiment score

s_s2 <- s_s %>%
  spread(sentiment, n, fill = 0) %>%
  mutate(sentiment_score = positive - negative)
```

```r
# now plot the sentiment scores over the progression of
# the books and story chunks

s_s2 %>%
  ggplot(aes(x = index, y = sentiment_score)) +
  geom_col(show.legend = F) +
  theme_bw()
```

```r
# let's plot a word cloud with sherlock holmes books
tidy_sherlock1 %>%
  count(word) %>%
  with(wordcloud(word, n, max.words = 100))
```

```r
# analysing word and document frequencies
# idf = ln(n_documents / n_documents_containing_term)
word_count_sherlock <- tidy_sherlock1 %>%
  count(word, sort = TRUE)
# create a new variable proportions
word_count_sh <- word_count_sherlock %>%
  mutate(proportion = n / sum(n))
# plot the two
word_count_sh %>%
  ggplot(aes(x = proportion)) +
  geom_histogram(show.legend = F) +
  xlim(NA, 0.0009) +
  theme_bw() +
  labs(title = "Distribution of words in Sherlock Stories",
      x = "Proportion of words",
      y = "Count of words")
```

```r
# many words occur rarely (left side)
# few words occur more frequently
# calculate frequency by rank

wc_rank <- word_count_sh %>%
  mutate(rank = row_number(),
        term_freq = proportion)
wc_rank %>%
  ggplot(aes(x = rank, y = term_freq)) +
  geom_line() +
  scale_x_log10() +
  scale_y_log10() +
  theme_bw() + 
  labs(title = "Term frequency by rank",
      x = "Rank",
      y = "Term frequency")
```
